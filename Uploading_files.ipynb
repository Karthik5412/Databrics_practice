## Schema Creation

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType

car_schema = StructType([StructField('car_id', StringType(), False),
                         StructField('car_name', StringType(), True),
                         StructField('year', IntegerType(), True),
                         StructField('price', FloatType(), True)])

                         
```

#### Using schema validating data

```python
df = spark.read.load("/Volumes/workspace/spark_practice/data/cars_perfect.csv",format='csv',schema=car_schema,header='true', inferschema='true')

df.show()
```

```python
df.display()
```

## Handling Corrupted Data

#### Skip bad records

```python
df = spark.read.format('csv')\
    .option('mode', 'PERMISSIVE')\
    .option('header', 'true')\
    .option('inferscahema', 'true')\
    .schema(car_schema)\
    .load('/Volumes/workspace/spark_practice/data/corrupted_car_data.csv')

df.display()
```

```python
df = spark.read.format('csv')\
    .option('mode', 'DROPMALFORMED')\
    .option('header', 'true')\
    .option('inferscahema', 'true')\
    .schema(car_schema)\
    .load('/Volumes/workspace/spark_practice/data/corrupted_car_data.csv')

df.display()
```

```python
df = spark.read.format('csv')\
    .option('mode', 'FASTFAIL')\
    .option('header', 'true')\
    .option('inferscahema', 'true')\
    .schema(car_schema)\
    .load('/Volumes/workspace/spark_practice/data/corrupted_car_data.csv')

df.display()
```

#### Storing Bad Records

```python
df = spark.read.format('csv')\
    .option('badRecordsPath', '/Volumes/workspace/spark_practice/data/corrupted_data')\
    .option('header', 'true')\
    .option('inferscahema', 'true')\
    .schema(car_schema)\
    .load('/Volumes/workspace/spark_practice/data/corrupted_car_data.csv')

df.display()
```
